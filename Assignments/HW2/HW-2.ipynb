**Question 4.** Briefly explain why a Latin Square Design would be appropriate here. Similarly, explain why a completely randomized design or randomized block design would not be appropriate.

 A Latin Square Design is appropriate because there are two known sources of variation (Day of the Week and Time of Day) that we want to control without inflating the number of observations. This design ensures that each platform appears exactly once in each row (Time of Day) and each column (Day), isolating the effect of the platform from the blocks.
A Completely Randomized Design would not control for Day or Time effects, potentially confounding results.
A Randomized Block Design controls for only one blocking factor, but we have two, making Latin Square the better choice.

Question 7. Build a three-way ANOVA model. Test the assumption of normality using both a visual method and a testing method. Do the results of the normality test(s) support the assumption of normality?

A two-way ANOVA showed that both Teaching Method and Prior Knowledge significantly affect quiz scores, while their interaction is not significant. The Q-Q plot showed residuals roughly followed a normal distribution, and the Shapiro-Wilk test returned a p-value greater than 0.05. Therefore, the normality assumption is reasonably satisfied for the ANOVA model.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import shapiro

# Load the dataset
file_path = '/content/STAT-7220-Applied-Experimental-Design/Assignments/HW2/Data Science Teaching Method.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Fit a two-way ANOVA model (with interaction)
model = ols('Score ~ C(Teaching_Method) * C(Prior_Knowledge)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("=== Two-Way ANOVA Table ===")
print(anova_table)

# Extract residuals
residuals = model.resid

# Q-Q Plot (Visual Normality Check)
sm.qqplot(residuals, line='45', fit=True)
plt.title('Q-Q Plot of ANOVA Residuals')
plt.tight_layout()
plt.show()

# Shapiro-Wilk Test (Statistical Normality Test)
shapiro_test = shapiro(residuals)
print("\n=== Shapiro-Wilk Test for Normality ===")
print(f"Test Statistic: {shapiro_test.statistic:.4f}")
print(f"p-value: {shapiro_test.pvalue:.4f}")

# Interpretation Guide (for your written answer)
if shapiro_test.pvalue > 0.05:
    print("\n✅ The residuals appear to be normally distributed (p > 0.05).")
else:
    print("\n❌ The residuals do not appear to be normally distributed (p < 0.05).")
```

Question 10. If the data more strongly support the alternative hypothesis, perform Tukey's HSD post-hoc test to determine which levels of the treatment effect are significantly different from each other. If the data more strongly support the null hypothesis, explain why a post-hoc test would not be appropriate.

Tukey's post-hoc test revealed that Direct Instruction significantly differs from both Collaborative Learning and Inquiry-Based Learning in terms of mean quiz scores. However, the difference between Collaborative Learning and Inquiry-Based Learning was not statistically significant. This suggests that Direct Instruction is less effective, while the other two methods perform similarly.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Run Tukey HSD for Teaching Method
tukey = pairwise_tukeyhsd(endog=df['Score'],
                          groups=df['Teaching_Method'],
                          alpha=0.05)

# Print results
print(tukey)

# Optional: Plot the Tukey HSD results
tukey.plot_simultaneous()
plt.title('Tukey HSD: Teaching Method Comparison')
plt.xlabel('Mean Difference in Scores')
plt.grid(True)
plt.tight_layout()
plt.show()
```
