7. Fit the full factorial ANOVA model. Check the assumption of normality using a testing method and a visual method. Does the assumption hold? Why or why not?

To test for differences in burrito satisfaction across ingredient combinations, I fit a full factorial ANOVA model including all main effects and interaction terms (Protein √ó Salsa √ó Potatoes). The ANOVA results showed statistically significant main effects for Protein (p = 0.025) and Salsa (p = 0.049), suggesting they both influence satisfaction. The main effect of Potatoes was marginal (p = 0.087), and all interaction effects were non-significant (p > 0.44), indicating that combinations of factors did not significantly affect ratings beyond the individual effects.
To check the assumption of normality of residuals, I used both a statistical and a visual method. The Shapiro-Wilk test gave a p-value of 0.135, which is greater than 0.05, so we fail to reject the null hypothesis and conclude the residuals are likely normally distributed. I also inspected the QQ plot, which showed that the residuals roughly followed the reference line, with only minor deviations at the tails. Based on both the test and the plot, the normality assumption appears to hold, and the ANOVA results can be considered valid.


```python
## Question 7 Code ##
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import shapiro
import matplotlib.pyplot as plt

# Ensure factors are categorical (in case not done yet)
df['Protein'] = df['Protein'].astype('category')
df['Salsa'] = df['Salsa'].astype('category')
df['Potatoes'] = df['Potatoes'].astype('category')

# --- Fit full factorial ANOVA model ---
model = ols('Satisfaction ~ Protein * Salsa * Potatoes', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("ANOVA Table:")
print(anova_table)

# --- Assumption Check: Normality of Residuals ---

# 1. Shapiro-Wilk Test
shapiro_test = shapiro(model.resid)
print("\nShapiro-Wilk Test for Normality:")
print(f"Test Statistic = {shapiro_test.statistic:.4f}, p-value = {shapiro_test.pvalue:.4f}")

# 2. QQ Plot
sm.qqplot(model.resid, line='s')
plt.title('QQ Plot of ANOVA Residuals')
plt.show()
```

8. Check the assumption of constant variance using a testing method and a visual method. Does the assumption hold? Why or why not?

To evaluate whether the assumption of constant variance (homoscedasticity) holds in the ANOVA model, I used both a statistical and a visual method. Levene‚Äôs test was applied across the 12 treatment groups, and the resulting p-value was above 0.05, indicating that we fail to reject the null hypothesis of equal variances. This suggests that the variance of residuals is reasonably consistent across groups. Additionally, the residuals vs. fitted values plot showed no clear fanning or funneling pattern. The spread of residuals appears relatively even across the range of fitted values, with no obvious trends or systematic structure. Together, the results of the test and the plot provide support for the constant variance assumption, meaning the ANOVA results can be interpreted with confidence in terms of this assumption.

```python
## Question 8 Code ##

from scipy.stats import levene
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Levene's Test for Homogeneity of Variance
# We group by the full combination of all three factors
grouped = df.groupby(['Protein', 'Salsa', 'Potatoes'])['Satisfaction'].apply(list)
levene_stat, levene_p = levene(*grouped)

print("Levene‚Äôs Test for Homogeneity of Variance:")
print(f"Test Statistic = {levene_stat:.4f}, p-value = {levene_p:.4f}")

# 2. Residuals vs. Fitted Plot
fitted_vals = model.fittedvalues
residuals = model.resid

plt.figure(figsize=(8, 6))
sns.scatterplot(x=fitted_vals, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Fitted Values (Homoscedasticity Check)')
plt.show()
```

Because the ANOVA results showed statistically significant main effects for both Protein and Salsa, I conducted a post‚Äëhoc analysis for the Protein factor, which has three levels and therefore requires pairwise comparisons to determine which specific proteins differ from one another. Using Tukey‚Äôs HSD test, I compared all pairs of protein types. The results showed that the only statistically significant difference was between Chorizo and Sausage (p = 0.0233), with Chorizo producing higher satisfaction ratings. 11. Calculate and interpret partial $\eta^2$ for the main effects.

To better understand the impact of each main effect, I calculated partial eta squared (Œ∑¬≤) values using the sums of squares from the ANOVA table. This effect size measure tells us the proportion of variance in satisfaction that each factor explains, after accounting for all other effects in the model. The partial Œ∑¬≤ for Protein was approximately 0.143, which is considered a large effect, meaning protein type had a substantial influence on how much people liked the burritos. The partial Œ∑¬≤ for Salsa was about 0.078, indicating a moderate effect, and for Potatoes, the value was around 0.060, also in the medium effect size range. These values support the earlier ANOVA results by showing that Protein and Salsa both have meaningful impacts on satisfaction, while Potatoes may still have a practical influence even if it wasn‚Äôt statistically significant at the 0.05 level.

```python
## Question 11 Code ##

# Calculate partial eta squared for each effect
# partial Œ∑¬≤ = SS_effect / (SS_effect + SS_residual)

# Extract sum of squares
ss_residual = anova_table.loc['Residual', 'sum_sq']

# Initialize empty dictionary to store effect sizes
partial_eta_squared = {}

# Loop through main effects only
main_effects = ['Protein', 'Salsa', 'Potatoes']
for effect in main_effects:
    ss_effect = anova_table.loc[effect, 'sum_sq']
    eta_sq = ss_effect / (ss_effect + ss_residual)
    partial_eta_squared[effect] = eta_sq

# Display results
print("Partial Œ∑¬≤ for Main Effects:")
for effect, eta in partial_eta_squared.items():
    print(f"{effect}: {eta:.4f}")
```

7. Assuming a defining relation of $I=ABC$, specify the aliasing structure of the design. What does this mean for the interpretation of the main effects and interactions?

In this 2¬≥‚Åª¬π fractional factorial design, the defining relation is
I=ABC.This creates an aliasing structure where each main effect is confounded with a two-factor interaction: specifically, A is aliased with BC, B with AC, and C with AB. This means that when we estimate the effect of a factor like Location (A), we are actually estimating the combined effect of Location and the interaction between Timing and Marketing (BC). The three-factor interaction ABC is aliased with the identity, meaning it cannot be separated from the general mean. As a result, while we can analyze and estimate effects, we must interpret them carefully. Any significant effect we find could be due to the main effect, the aliased interaction, or both ‚Äî we can‚Äôt tell them apart. This is a trade-off of using a fractional design, and we generally assume higher-order interactions are small enough to ignore.

```python
## Question 8 Code ##

import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure categorical types
df['Location'] = df['Location'].astype('category')
df['Timing'] = df['Timing'].astype('category')
df['Marketing'] = df['Marketing'].astype('category')

# Fit the ANOVA model
model = smf.ols('Sales ~ Location + Timing + Marketing', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display ANOVA table
print("ANOVA Table:")
print(anova_table)

# --- Normality Checks ---

# Extract residuals
residuals = model.resid

# Shapiro-Wilk Test
shapiro_stat, shapiro_p = shapiro(residuals)
print("\nShapiro-Wilk Test for Normality:")
print(f"Test Statistic = {shapiro_stat:.4f}, p-value = {shapiro_p:.4f}")

# Q-Q Plot
sm.qqplot(residuals, line='s')
plt.title("Q-Q Plot of Residuals (Normality Check)")
plt.show()

# Histogram (optional, for additional visual check)
plt.figure(figsize=(6, 4))
sns.histplot(residuals, kde=True)
plt.title("Histogram of Residuals")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

The p-value was greater than 0.05, which means we fail to reject the null hypothesis and have no strong evidence of unequal variances. Together, these results support the conclusion that the assumption of constant variance is reasonably met in this model.


```python
## Question 9 Code ##

from scipy.stats import levene

# --- 1. Residuals vs. Fitted Plot ---
fitted = model.fittedvalues
residuals = model.resid

plt.figure(figsize=(8, 6))
plt.scatter(fitted, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuals vs. Fitted Values (Homoscedasticity Check)")
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.grid(True)
plt.show()

# --- 2. Levene's Test for Equal Variance ---
# Group by each unique treatment combination
df['residuals'] = residuals
groups = df.groupby(['Location', 'Timing', 'Marketing'])['residuals'].apply(list)

# Perform Levene's test on the residual groups
levene_stat, levene_p = levene(*groups)
print("Levene‚Äôs Test for Homogeneity of Variance:")
print(f"Test Statistic = {levene_stat:.4f}, p-value = {levene_p:.4f}")
```

The ANOVA table reveals that all three main effects‚ÄîLocation, Timing, and Marketing‚Äîhave statistically significant impacts on sales.
```python
## Question 10 Code ##

import statsmodels.api as sm
import statsmodels.formula.api as smf

# Ensure all factors are treated as categorical
df['Location'] = df['Location'].astype('category')
df['Timing'] = df['Timing'].astype('category')
df['Marketing'] = df['Marketing'].astype('category')

# Fit ANOVA model with main effects only (since it's a 2^3-1 design)
model = smf.ols('Sales ~ Location + Timing + Marketing', data=df).fit()

# Get ANOVA table (Type II sums of squares)
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print("ANOVA Table:")
print(anova_table)
```

11. Calculate and interpret the values for the main effects. Which effects seem more important? Why?

The main effects help us understand how each factor individually impacts sales. For Location, the main effect is ‚Äì117.06, meaning that sales were about $117 lower on average at the University compared to the Brewery. For Timing, the effect is +343.49, suggesting that setting up on weekends rather than weekdays increases average sales by roughly $343. Marketing also shows a strong effect of +245.97, meaning that using partnership marketing (with the brewery or university) increases sales compared to independent efforts. Among these, Timing has the largest impact, followed by Marketing, with Location having the smallest (and negative) effect. This suggests that focusing on weekend setups and strategic partnerships is likely the best way to boost sales.


```python
# Map levels to -1 and +1 for clarity
df['A'] = df['Location'].map({'Brewery': -1, 'University': 1})
df['B'] = df['Timing'].map({'Weekday': -1, 'Weekend': 1})
df['C'] = df['Marketing'].map({'Independent': -1, 'Partnership': 1})

# Calculate main effects
def main_effect(df, factor):
    high = df[df[factor] == 1]['Sales'].mean()
    low = df[df[factor] == -1]['Sales'].mean()
    return high - low

effect_A = main_effect(df, 'A')  # Location
effect_B = main_effect(df, 'B')  # Timing
effect_C = main_effect(df, 'C')  # Marketing

print(f"Main Effect of Location (A): {effect_A:.2f}")
print(f"Main Effect of Timing (B): {effect_B:.2f}")
print(f"Main Effect of Marketing (C): {effect_C:.2f}")
```

12. Calculate and interpret the partial $\eta^2$ for the main effects. Which effects seem more important? Why?

To assess the relative importance of each factor, I calculated the partial Œ∑¬≤ for Location, Timing, and Marketing. These values estimate how much variance in sales can be explained by each factor, after accounting for the others. The partial Œ∑¬≤ for Timing was 0.839, meaning it explains nearly 84% of the model‚Äôs explained variance, which is a very large effect. Marketing had a partial Œ∑¬≤ of 0.727, also a very large effect. Location had a partial Œ∑¬≤ of 0.376, still a large effect, but clearly smaller than the others. These results reinforce the earlier conclusions: Timing and Marketing are the most influential factors, with Timing having the greatest impact on sales.

```python
## Question 12 Code ##

# Calculate partial eta squared for each main effect
# Formula: Œ∑¬≤_partial = SS_effect / (SS_effect + SS_residual)

# Extract SS values from ANOVA table
ss_residual = anova_table.loc['Residual', 'sum_sq']

partial_eta_squared = {}
for effect in ['Location', 'Timing', 'Marketing']:
    ss_effect = anova_table.loc[effect, 'sum_sq']
    eta_sq = ss_effect / (ss_effect + ss_residual)
    partial_eta_squared[effect] = eta_sq

# Print the results
print("Partial Œ∑¬≤ for Main Effects:")
for effect, eta in partial_eta_squared.items():
    print(f"{effect}: {eta:.4f}")
```
