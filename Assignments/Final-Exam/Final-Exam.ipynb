## Part 1 Instructions:
Let's say you own a new food truck called **Buster's Breakfast Burritos**. To determine the best ingredients for your burritos, and as a means of minimizing food (and money) waste, you decide to conduct an experiment. Specifically, you want to investigate the following treatments:

| Factor   | Levels                  |
|:---------|:------------------------|
| Protein  | Bacon, Sausage, Chorizo |
| Salsa    | Mild, Hot               |
| Potatoes | Yes, No                 |

To gather data for this experiment, you recruit $r=5$ local residents for each of the $3\times 2\times 2=12$ treatment combinations. After each participant has tasted their randomly assigned burrito, you ask them to rate the burrito on a scale of 1 to 10 (1 = terrible, 10 = amazing). The data are stored in the file `Busters Burritos.xlsx` in the assignment repo. With these data, I want you to:

1. Briefly state the objective of the experiment.
The goal of this experiment is to figure out which combination of ingredients makes the best breakfast burrito according to local taste preferences. By testing different combinations of proteins, salsa types, and whether or not potatoes are included, I want to find the best-tasting burrito while also minimizing ingredient waste and saving money.


2. State the outcome variable and how it was measured.

The outcome variable is the burrito taste rating, which was measured on a scale from 1 to 10. Each participant gave a score after trying one burrito, where 1 means it tasted terrible and 10 means it was amazing. This gives us a numerical way to compare how each combination of ingredients is perceived.

3. State the main effects. What lurking variables may be present?

The main effects in this experiment are the individual effects of:
Protein (Bacon, Sausage, Chorizo)
Salsa (Mild, Hot)
Potatoes (Yes, No)
Each main effect tells us how changing just one factor (e.g., switching from bacon to chorizo) impacts the average burrito rating, regardless of the other ingredients.
Possible lurking variables (hidden influences) could include:
The time of day the burrito was eaten (breakfast vs. lunch)
Whether the participant was hungry or already full
Personal preferences not accounted for (e.g., someone hates spicy food)
The way the burrito was prepared or served (maybe one cook made it better than another)
Order effects if someone tried more than one burrito

4. Why might a full factorial design be preferred here to a $2^k$ fractional factorial design?

A full factorial design tests every possible combination of the three factors and their levels (3 proteins √ó 2 salsa levels √ó 2 potato options = 12 combinations). This is preferred here because:
We're not just interested in which individual factor is best ‚Äî we also want to see interactions (e.g., maybe chorizo only works well with hot salsa).
Since there are only 12 combinations and we can easily gather 5 ratings for each, it‚Äôs totally doable and practical to run the full experiment.
A fractional factorial (like 2¬≥) would cut out some combinations, which could mean missing key insights about what people really like.


5. State all sets of statistical hypotheses.

 Main Effect of Protein
Null (H‚ÇÄ): The mean burrito ratings are the same for bacon, sausage, and chorizo.
Alternative (H‚ÇÅ): At least one protein leads to a different mean burrito rating.
Main Effect of Salsa
H‚ÇÄ: The mean burrito ratings are the same for mild and hot salsa.
H‚ÇÅ: The mean burrito ratings differ between mild and hot salsa.
 Main Effect of Potatoes
H‚ÇÄ: The mean ratings are the same whether potatoes are included or not.
H‚ÇÅ: Including potatoes changes the average burrito rating.

Interaction Effects
For each pair or trio of factors, we also test whether the combination of ingredients matters more than the ingredients individually.
Protein √ó Salsa Interaction
H‚ÇÄ: No interaction ‚Äî the effect of protein does not depend on salsa type.
H‚ÇÅ: There is an interaction ‚Äî the best protein depends on the salsa used.
Protein √ó Potatoes Interaction
H‚ÇÄ: No interaction between protein and potatoes.
H‚ÇÅ: The effect of protein depends on whether potatoes are included.
Salsa √ó Potatoes Interaction
H‚ÇÄ: No interaction between salsa and potatoes.
H‚ÇÅ: The effect of salsa depends on the presence of potatoes.
Protein √ó Salsa √ó Potatoes (3-way interaction)
H‚ÇÄ: No three-way interaction.
H‚ÇÅ: The effect of one factor depends on the specific combination of the other two.

Chorizo and mild salsa were generally associated with higher ratings, while sausage and hot salsa tended to score lower. Potatoes slightly improved ratings overall.
```python
## Question 6 Code ##
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Updated file path
file_path = '/content/STAT-7220-Applied-Experimental-Design/Assignments/Final-Exam/Busters Burritos.xlsx'

# Load the data
df = pd.read_excel(file_path)

# Preview the data
print(df.head())

# Convert factors to categorical
df['Protein'] = df['Protein'].astype('category')
df['Salsa'] = df['Salsa'].astype('category')
df['Potatoes'] = df['Potatoes'].astype('category')

# Summary statistics
summary = df.groupby(['Protein', 'Salsa', 'Potatoes'])['Satisfaction'].agg(['mean', 'std', 'count', 'min', 'max']).reset_index()
print("\nSummary Statistics:")
print(summary)

# Boxplots for main effects
plt.figure(figsize=(10, 6))
sns.boxplot(x='Protein', y='Satisfaction', data=df)
plt.title('Burrito Rating by Protein')
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x='Salsa', y='Satisfaction', data=df)
plt.title('Burrito Rating by Salsa')
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x='Potatoes', y='Satisfaction', data=df)
plt.title('Burrito Rating by Potatoes')
plt.show()

# Interaction plot: Protein √ó Salsa
fig = plt.figure(figsize=(8, 6))
sm.graphics.interaction_plot(df['Protein'].astype(str), df['Salsa'].astype(str), df['Satisfaction'], markers=['D','^'], ms=10)
plt.title('Interaction Plot: Protein √ó Salsa')
plt.show()

# Interaction plot: Protein √ó Potatoes
fig = plt.figure(figsize=(8, 6))
sm.graphics.interaction_plot(df['Protein'].astype(str), df['Potatoes'].astype(str), df['Satisfaction'], markers=['D','o'], ms=10)
plt.title('Interaction Plot: Protein √ó Potatoes')
plt.show()

# Interaction plot: Salsa √ó Potatoes
fig = plt.figure(figsize=(8, 6))
sm.graphics.interaction_plot(df['Salsa'].astype(str), df['Potatoes'].astype(str), df['Satisfaction'], markers=['D','o'], ms=10)
plt.title('Interaction Plot: Salsa √ó Potatoes')
plt.show()
```
