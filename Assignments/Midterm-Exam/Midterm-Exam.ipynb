Question 6. Perform appropriate exploratory analysis, including summary statistics and data visualizations. Do the results of these analyses support the null or alternative hypothesis more strongly?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Load the data
file_path = "/content/STAT-7220-Applied-Experimental-Design/Assignments/Midterm-Exam/Video Game Controllers.xlsx"
video_game_df = pd.read_excel(file_path)

# -----------------------------
# Q6: Summary statistics & visualization
# -----------------------------
summary_stats = video_game_df.groupby('Material')['Comfort'].describe()
print("\nSummary Statistics by Material:")
print(summary_stats)

# Boxplot with individual data points
plt.figure(figsize=(8, 6))
sns.boxplot(data=video_game_df, x='Material', y='Comfort')
sns.stripplot(data=video_game_df, x='Material', y='Comfort', color='black', alpha=0.6, jitter=True)
plt.title('Comfort Ratings by Grip Material')
plt.ylabel('Comfort (1-10)')
plt.xlabel('Material')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Question 7. Build a two-way ANOVA model. Test the assumption of normality using both a visual method and a testing method. Do the results of the normality test(s) support the assumption of normality?

The normality of residuals was tested using both a Q-Q plot and the Shapiro-Wilk test. The Q-Q plot shows that the residuals closely follow the red reference line, indicating that they are approximately normally distributed. The Shapiro-Wilk test produced a p-value of 0.951, which is greater than 0.05, so we fail to reject the null hypothesis of normality. Therefore, both the visual and statistical methods suggest that the assumption of normality is met for this ANOVA model.

```python
# -----------------------------
# Q7: One-way ANOVA + Normality test
# -----------------------------
# Fit the ANOVA model
model = ols('Comfort ~ C(Material)', data=video_game_df).fit()

# Residual normality test (Shapiro-Wilk)
shapiro_test = stats.shapiro(model.resid)
print("Shapiro-Wilk Test:", shapiro_test)

# Q-Q Plot
sm.qqplot(model.resid, line='s')
plt.title("Q-Q Plot of Residuals")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Question 8. Test the assumption of homogeneity of variance using both a testing and visual method. Do the results of the test(s) support the assumption of homogeneity of variance?

Levene‚Äôs test was used to check the assumption of equal variances among the four material groups. The test produced a statistic of 1.3685 with a p-value of 0.2679. Since the p-value is greater than 0.05, we fail to reject the null hypothesis that the group variances are equal. In addition, the residuals versus fitted values plot showed no clear pattern or funnel shape, further supporting this result. Therefore, the assumption of homogeneity of variance is satisfied for this ANOVA model.

```python
# -----------------------------
# Q8: Homogeneity of variances test
# -----------------------------
# Levene‚Äôs test
levene_test = stats.levene(
    *[group["Comfort"].values for name, group in video_game_df.groupby("Material")]
)
print("\nLevene‚Äôs Test for Equal Variance:")
print(f"Statistic = {levene_test.statistic:.4f}, p-value = {levene_test.pvalue:.4f}")

# Residuals vs Fitted Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=model.fittedvalues, y=model.resid)
plt.axhline(0, linestyle='--', color='gray')
plt.title("Residuals vs Fitted Values")
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Question 9. Report the F-statistic and its associated p-value for the test for the treatment factor. Which of our two hypotheses is more strongly supported? Why?

The ANOVA test produced an F-statistic of 47.48 with a corresponding p-value of 1.35 √ó 10‚Åª¬π¬≤. Since the p-value is far below the common significance level of 0.05, we reject the null hypothesis that all group means are equal. This result provides strong statistical evidence that the type of grip material has a significant effect on the comfort rating. Therefore, the data more strongly support the alternative hypothesis: at least one material differs in mean comfort level from the others.

```python
# -----------------------------
# Q9: ANOVA Table
# -----------------------------
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nANOVA Table:")
print(anova_table)
```

**Question 10.** If the data more strongly support the alternative hypothesis, perform Tukey's HSD post-hoc test to determine which levels of the treatment effect are significantly different from each other. If the data more strongly support the null hypothesis, explain why a post-hoc test would not be appropriate.


Tukey‚Äôs HSD post-hoc test was conducted following a significant ANOVA result to determine which grip materials differed in average comfort ratings. The test showed that Silicone was significantly more comfortable than Metal, Plastic, and Rubber. Metal was significantly less comfortable than all other materials. However, there was no significant difference between Plastic and Rubber (p = 0.1104). These results suggest that while some materials (like Silicone and Metal) stand out in terms of comfort, others (Plastic and Rubber) are perceived similarly by users.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Run Tukey's HSD test
tukey_result = pairwise_tukeyhsd(
    endog=video_game_df['Comfort'],
    groups=video_game_df['Material'],
    alpha=0.05
)

# Print the results
print(tukey_result.summary())
```
